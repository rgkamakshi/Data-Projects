```python
from matplotlib import pyplot as plt
import pandas as pd
df = pd.read_csv('insurance.csv')
df.head()
```

```python
x = df['bmi']
y = df['charges']
plt.scatter(x,y,c=df['smoker'],alpha = 0.7)
plt.xlabel('BMI')
plt.ylabel('Medical costs')
plt.title('BMI vs Medical costs')
#plt.legend(['workingday','nonworkingday'])
cbar = plt.colorbar(ticks = [1,0])
cbar.set_ticklabels(['Smoking','Non smoking'])
plt.show()
```

For non-smokers - We can see that for non-smokers the medical cost is mostly within 15,000 irrespective of the BMI with a few outliers.
For smokers - Medical costs for smokers are comparatively higher than that of non-smokers. The cost increases with the increase in BMI.

```python
smokers = df.loc[ df.loc[ :, 'smoker'] == 1 ,'charges']
non_smokers = df.loc[ df.loc[ :, 'smoker'] == 0 ,'charges']
plt.hist(smokers,12,alpha = 1)
plt.hist(non_smokers,12,alpha = 0.6)
plt.xlabel("Medical cost")
plt.ylabel("No of observation")
plt.title("Distribution of medical costs of Smokers vs Non-smokers ")
plt.legend(["Smoker", "Non-smoker"])
plt.show()
```

For non-smokers:
	As discussed in the previous question, the medical cost of non-smokers is lower than that of smokers.
	Medical cost for non-smokers is not normally distributed
	The medical cost for non-smokers is skewed towards the left i.e., for a larger number of observations, medical cost is lower
For smokers:
	The medical cost for non-smokers is mostly within 15,000 whereas for smokers, the cost ranges anywhere between 15,000-60,000
	For majority of the observation, the medical costs seem to be around 20,000 and between 35,000-55,000.

```python
younger_ppl = df.loc[ df.loc[ :, 'age'] < 40 ,'charges']
older_ppl = df.loc[ df.loc[ :, 'age'] > 40 ,'charges']
plt.hist(younger_ppl,15)
plt.hist(older_ppl,15,alpha = 0.7)
plt.xlabel("Medical cost")
plt.ylabel("No of observation")
plt.title("Distribution of medical cost of young people and that of elder people")
plt.legend(["Young people (age < 40)", "Older people (age > 40)"])
plt.show()
```

Insights: 
	The distribution is not normal for both younger and older people.
	For majority of the younger population, the medical cost is lesser than 10,000.
	For the older population, the medical cost is distributed between 5000 and 55000.
	In both the younger and older population, the data is skewed towards the left i.e., for a larger number of observations, medical cost is lower.

```python
smokers = df.loc[ df.loc[ :, 'smoker'] == 1 ,'region']
non_smokers = df.loc[ df.loc[ :, 'smoker'] == 0 ,'region']
plt.hist(smokers,alpha = 1,color = 'blue')
plt.hist(non_smokers,alpha = 0.6)
plt.xticks( [0,1,2,3], ["Southwest", "Southeast", 'Northwest', 'Northeast'])
plt.xlabel("Region")
plt.ylabel("No of observation")
plt.title("Smokers vs Non smokers by region ")
plt.legend(['Smokers','Non smokers'])
plt.show()
```

Insights: From the histogram, it can be inferred that the number of smokers in the southeast region are comparatively higher than the other three regions which leads us to the next question if this increase has an impact on the medical cost in that region.

```python
southwest = df.loc[ df.loc[ :, 'region'] == 0 ,'charges']
southeast = df.loc[ df.loc[ :, 'region'] == 1 ,'charges']
northwest = df.loc[ df.loc[ :, 'region'] == 2 ,'charges']
northeast = df.loc[ df.loc[ :, 'region'] == 3 ,'charges']

plt.hist(southwest, 10,alpha = 0.8)
plt.hist(southeast, 10, alpha = 0.8)
plt.hist(northwest, 10, alpha = 0.8)
plt.hist(northeast, 10, alpha = 0.7)
plt.xlabel("Medical cost by region")
plt.ylabel("No of observations")
plt.legend(["Southwest", "Southeast", 'Northwest', 'Northeast'])
plt.title('Distribution of medical cost by region')
plt.show()
```

```python
plt.boxplot( [southwest, southeast, northwest, northeast], showmeans = True)
plt.xticks( [1,2,3,4], ["Southwest", "Southeast", 'Northwest', 'Northeast'])
plt.ylabel('charges')
plt.show()
```

Insights : As expected it can be seen that southeast has slightly higher medical costs compared to the other regions which can be tied to our previous inference ,i.e high number of smokers in that area
