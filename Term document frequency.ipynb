```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

```python
df=pd.read_csv('IA2.csv',header=None)
df.set_index(df.columns[0],inplace=True)
df.head()
```

```python
# 1) Tokenize each review in the collection; 

all_tokens=[]
for i in df[1]:
    tokens = word_tokenize(i.lower())
    all_tokens.append(tokens)
print(all_tokens)
```

```python
#2) Use the tokenized reviews after step 1, lemmatize all the words;

lemmatizer = nltk.stem.WordNetLemmatizer()
lemmatized_tokens = []
for tokens in all_tokens:
    lemmatized_tokens.append([lemmatizer.lemmatize(token) for token in tokens if token.isalpha()])
print(lemmatized_tokens)
```

```python
#3)  Based on the output in step 2, remove all the stop‚Äêwords and the punctuations;
#----Removed punctuation in last step when checking for isalpha.

stop_words = set(stopwords.words('english'))
stop_words_removed = [[t for t in review if not t in stop_words] for review in lemmatized_tokens]

#print(stop_words_removed)
```

```python
docs = [' '.join(tokens) for tokens in stop_words_removed]
print(docs)
```

```python
# 4)  Based on the output in step 3, convert each of the reviews in TD‚ÄêIDF vectors. The minimal 
# document frequency for each term is 3. Also, include 2‚Äêgram; 

###TF-IDF
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer1 = TfidfVectorizer(ngram_range=(1,2), min_df=3)
vectorizer1.fit(docs)
print(vectorizer1.vocabulary_)

v1 = vectorizer1.transform(docs)
td_idf=v1.toarray()
df_new = pd.DataFrame(td_idf,columns=vectorizer1.get_feature_names())
df_new = df_new.reset_index(drop=True)
df_new.index = df_new.index + 1

print(df_new)
print(df_new.shape)
```

```python
#concating the new dataframe that has td-idf vectors with original data frame that has the 100 reviews

result = pd.concat([df, df_new], axis=1)
print(result)
print(result.shape)
```

```python
#Saving to csv

result.to_csv('Part1_Q4.csv', index=False)
```

```python
# 5)  Based on the output in step 1, POS‚Äêtag each word and do a TD‚ÄêIDF vectorization, the minimal 
#document frequency for each term is 4 (please don‚Äôt do normalization and stop‚Äêword removal)

POS_c2 = []
for doc in all_tokens:
    word_list=[]
    for token in doc:
        if token.isalpha():
            word_list.append(token)
    #token_doc = nltk.word_tokenize(doc)
    POS_token_doc = nltk.pos_tag(word_list)
    POS_token_temp = []
    for i in POS_token_doc:
         POS_token_temp.append(i[0] + "_" + i[1])
    POS_c2.append(" ".join(POS_token_temp))         
    
print(POS_c2)
```

```python
###TF-IDF
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer2 = TfidfVectorizer(min_df=4)
vectorizer2.fit(POS_c2)
print(vectorizer2.vocabulary_)
v2 = vectorizer2.transform(POS_c2)
pos=v2.toarray()

pos_df = pd.DataFrame(pos,columns=vectorizer2.get_feature_names())
pos_df = pos_df.reset_index(drop=True)
pos_df.index = pos_df.index + 1
```

```python
#concating the new dataframe that has td-idf vectors for pos tag with original data frame that has the 100 reviews

result_pos = pd.concat([df, pos_df], axis=1)
print(result_pos)
print(result_pos.shape)
```

```python
#Saving to csv


result_pos.to_csv('Part1_Q5.csv', index=False)
```
